#**Cek Outlier**

##**Interquartile Range (IQR)**

```python
def cari_outlier(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   jumlah = ((df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR))).sum()
   letak = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return jumlah, letak
```

```python
jumlah = cari_outlier(df)[0]
jumlah 
```

```python
letak = cari_outlier(df)[1]
letak 
```

Pada dataframe di atas, data yang masih berupa angka merupakan data outlier, sedangkan data NaN bukan data outlier

#**Cek Distribusi**

##**Kolmogorov Smirnov**

```python
from scipy.stats import kstest
import scipy.stats as stats
from scipy.stats import normaltest
```

```python
#H0 = Data Berdistribusi Normal
#H1 = Data Tidak Beridstribusi Normal
def check_p_val(p_val, alpha):

    if p_val < alpha:
        print('H0 ditolak, data tidak berdistribusi normal')
    else:
        print('H0 diterima, data berdistribusi normal')
```

```python
#cek normalitas menggunakan kolmogorov smirnov
stat, p_val = kstest(df['price'], 'norm')
print('Statistic: \t{:1.2f} \nP-Value: \t{:1.2e}\n'.format(stat, p_val))
check_p_val(p_val, alpha=0.05)
```

```python
#cek normalitas menggunakan kolmogorov smirnov
stat, p_val = kstest(df['brand'], 'norm')
print('Statistic: \t{:1.2f} \nP-Value: \t{:1.2e}\n'.format(stat, p_val))
check_p_val(p_val, alpha=0.05)
```

##**Spearman**

```python
from scipy.stats import spearmanr
```

```python
c = df.columns.values
c
```

```python
Y, X1, X2, X3, X4, X5, X6, X7, X8, X9, X10, X11 = df[c[0]], df[c[1]], df[c[2]], df[c[3]], df[c[4]], df[c[5]], df[c[6]], df[c[7]], df[c[8]], df[c[9]], df[c[10]], df[c[11]]
```

```python
corr, _ = spearmanr(X1, Y)
print('Spearmans correlation brand vs price: %.3f' % corr)

corr, _ = spearmanr(X2, Y)
print('Spearmans correlation model vs price: %.3f' % corr)

corr, _ = spearmanr(X3, Y)
print('Spearmans correlation year vs price: %.3f' % corr)

corr, _ = spearmanr(X4, Y)
print('Spearmans correlation title_status vs price: %.3f' % corr)

corr, _ = spearmanr(X5, Y)
print('Spearmans correlation mileage vs price: %.3f' % corr)

corr, _ = spearmanr(X6, Y)
print('Spearmans correlation color vs price: %.3f' % corr)

corr, _ = spearmanr(X7, Y)
print('Spearmans correlation vin vs price: %.3f' % corr)

corr, _ = spearmanr(X8, Y)
print('Spearmans correlation lot vs price: %.3f' % corr)

corr, _ = spearmanr(X9, Y)
print('Spearmans correlation state vs price: %.3f' % corr)

corr, _ = spearmanr(X10, Y)
print('Spearmans correlation country vs price: %.3f' % corr)

corr, _ = spearmanr(X11, Y)
print('Spearmans correlation condition vs price: %.3f' % corr)
```

Jika melihat skor spearman, dapat dilihat bahwa brand, title_status, mileage, vin, dan country memiliki korelasi negatif terhadap variabel price, dan strong negatif terhadap variabel price adalah variabel vin, sisanya memiliki korelasi positif terhadap variabel price [[kutipan]](https://stackabuse.com/calculating-pearson-correlation-coefficient-in-python-with-numpy/)
