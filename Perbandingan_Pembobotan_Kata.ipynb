```python
import pandas as pd
```

#**Import Data**

```python
df = pd.read_excel('Text Clean dan Label Lexicon.xlsx').dropna()
df.head()
```

##**Membagi Data Menjadi Train dan Test**

```python
X = df['Text_Clean']
y = df['polarity']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 20)
```

```python
X_train
```

```python
y_train
```

##**LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
y_train = Encoder.fit_transform(y_train)
y_test = Encoder.fit_transform(y_test)
```

#**Vector TF**

```python
from sklearn.feature_extraction.text import CountVectorizer
Tf = CountVectorizer()
Tf.fit(df['Text_Clean'])
X_train_Tf = Tf.transform(X_train)
X_test_Tf = Tf.transform(X_test)
```

```python
X_train_Tf.toarray()
```

```python
df_tf_train = pd.DataFrame(X_train_Tf.toarray(), columns=Tf.get_feature_names_out(), index=X_train)
df_tf_train.head()
```

```python
df_tf_test = pd.DataFrame(X_test_Tf.toarray(), columns=Tf.get_feature_names_out(), index=X_test)
df_tf_test.head()
```

```python
print('Dimensi data X_train sebelum TF :' ,X_train.shape)
print('Dimensi data X_train sesudah TF :' ,df_tf_train.shape)
```

```python
print('Dimensi data X_test sebelum TF :' ,X_test.shape)
print('Dimensi data X_test sesudah TF :' ,df_tf_test.shape)
```

#**Vector TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf = TfidfVectorizer().fit(df['Text_Clean'])
X_train_Tfidf = Tfidf.transform(X_train)
X_test_Tfidf = Tfidf.transform(X_test)
```

```python
X_train_Tfidf.toarray()
```

```python
df_tfidf_train = pd.DataFrame(X_train_Tfidf.toarray(), columns=Tfidf.get_feature_names(), index=X_train)
df_tfidf_train.head()
```

```python
df_tfidf_test = pd.DataFrame(X_test_Tfidf.toarray(), columns=Tfidf.get_feature_names(), index=X_test)
df_tfidf_test.head()
```

```python
print('Dimensi data X_train sebelum TF-IDF :' ,X_train.shape)
print('Dimensi data X_train sesudah TF-IDF :' ,df_tfidf_train.shape)
```

```python
print('Dimensi data X_test sebelum TF-IDF :' ,X_test.shape)
print('Dimensi data X_test sesudah TF-IDF :' ,df_tfidf_test.shape)
```

#**Delta TF-IDF**

```python
from sklearn_deltatfidf import DeltaTfidfVectorizer
delta = DeltaTfidfVectorizer()
delta.fit_transform(df['Text_Clean'], list(y))

X_train_delta = delta.transform(X_train)
X_test_delta = delta.transform(X_test)
```

```python
X_train_delta.toarray()
```

```python
df_delta_train = pd.DataFrame(X_train_delta.toarray(), columns=delta.get_feature_names_out(), index=X_train)
df_delta_train.head()
```

```python
df_delta_test = pd.DataFrame(X_test_delta.toarray(), columns=delta.get_feature_names_out(), index=X_test)
df_delta_test.head()
```

```python
print('Dimensi data X_train sebelum Delta TF-IDF :' ,X_train.shape)
print('Dimensi data X_train sesudah Delta TF-IDF :' ,df_delta_train.shape)
```

```python
print('Dimensi data X_test sebelum Delta TF-IDF :' ,X_test.shape)
print('Dimensi data X_test sesudah Delta TF-IDF :' ,df_delta_test.shape)
```

#**Support Vector Machine**

##**Menggunakan TF**

```python
#klasifikasi naive bayes
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

clf = SVC().fit(X_train_Tf, y_train)
pred_tf = clf.predict(X_test_Tf)
print("MultinomialNB Accuracy:" , accuracy_score(y_test,pred_tf))
print(classification_report(y_test,pred_tf,zero_division=0))
```

##**Menggunakan TF-IDF**

```python
#klasifikasi naive bayes
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

clf = SVC().fit(X_train_Tfidf, y_train)
pred_idf = clf.predict(X_test_Tfidf)
print("MultinomialNB Accuracy:" , accuracy_score(y_test,pred_idf))
print(classification_report(y_test,pred_idf,zero_division=0))
```
