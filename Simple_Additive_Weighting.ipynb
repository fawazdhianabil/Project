```python
def SAW():
  print('-'*20)
  print('Variable Benefit')
  print('Program Simple Adiditive Weighted untuk Pelamar Kerja')
  print('Pengalaman Kerja (C1)')
  print('Pendidikan (C2)')
  print('Usia (C3)')
  print('Variabel Cost')
  print('Status Perkawinan (C4)')
  print('Alamat (C5)')
  print('-'*20)
  print('Calon Pegawai')
  print('Calon Pegawai 1 (A1)')
  print('Calon Pegawai 2 (A2)')
  print('Calon Pegawai 3 (A3)')
  print('Calon Pegawai 4 (A4)')
  print('Calon Pegawai 5 (A5)')

  #input nilai kriteria
  c1_A1, c1_A2, c1_A3, c1_A4, c1_A5 = [float(s) for s in input('Masukkan 5 nilai C1 dengan Spasi: ').split()]
  c2_A1, c2_A2, c2_A3, c2_A4, c2_A5 = [float(s) for s in input('Masukkan 5 nilai C2 dengan Spasi: ').split()]
  c3_A1, c3_A2, c3_A3, c3_A4, c3_A5 = [float(s) for s in input('Masukkan 5 nilai C3 dengan Spasi: ').split()]
  c4_A1, c4_A2, c4_A3, c4_A4, c4_A5 = [float(s) for s in input('Masukkan 5 nilai C4 dengan Spasi: ').split()]
  c5_A1, c5_A2, c5_A3, c5_A4, c5_A5 = [float(s) for s in input('Masukkan 5 nilai C5 dengan Spasi: ').split()]


  #menampilkan kriteria para pegawai
  print('-'*20)
  import pandas as pd
  print('Kriteria Para Calon Pegawai')
  df = pd.DataFrame({'Calon Pegawai':['A1','A2','A3','A4','A5'],
                    'C1':[c1_A1,c1_A2,c1_A3,c1_A4,c1_A5],
                    'C2':[c2_A1,c2_A2,c2_A3,c2_A4,c2_A5],
                    'C3':[c3_A1,c3_A2,c3_A3,c3_A4,c3_A5],
                    'C4':[c4_A1,c4_A2,c4_A3,c4_A4,c4_A5],
                    'C5':[c5_A1,c5_A2,c5_A3,c5_A4,c5_A5]})
  df1 = df.set_index(df['Calon Pegawai']).drop(columns=['Calon Pegawai'])
  print(df1)
  print('-'*20)

  #mendefinisikan bobot
  w1,w2,w3,w4,w5 = [float(s) for s in input('Masukkan 5 nilai Bobot dengan Spasi: ').split()]

  #menampilkan nilai bobot
  print('-'*20)
  print('Nilai Bobot Calon Pegawai Pertama',w1)
  print('Nilai Bobot Calon Pegawai Kedua',w2)
  print('Nilai Bobot Calon Pegawai Ketiga',w3)
  print('Nilai Bobot Calon Pegawai Keempat',w4)
  print('Nilai Bobot Calon Pegawai Kelima',w5)
  print('-'*20)

  #mencari nilai maksimum pada kategori benefit
  import numpy as np
  c1_max = [np.max(i) for i in df[:1].drop(columns=['Calon Pegawai']).values][0]
  c2_max = [np.max(i) for i in df[1:2].drop(columns=['Calon Pegawai']).values][0]
  c3_max = [np.max(i) for i in df[2:3].drop(columns=['Calon Pegawai']).values][0]

  #mencari nilai minimum pada kategori cost
  c4_min = [np.min(i) for i in df[3:4].drop(columns=['Calon Pegawai']).values][0]
  c5_min = [np.min(i) for i in df[4:5].drop(columns=['Calon Pegawai']).values][0]

  #mencari nilai normalisasi R1
  R11 = c1_A1/c1_max
  R12 = c1_A2/c1_max
  R13 = c1_A3/c1_max
  R14 = c1_A4/c1_max
  R15 = c1_A5/c1_max

  #mencari nilai normalisasi R2
  R21 = c2_A1/c2_max
  R22 = c2_A2/c2_max
  R23 = c2_A3/c2_max
  R24 = c2_A4/c2_max
  R25 = c2_A5/c2_max

  #mencari nilai normalisasi R3
  R31 = c3_A1/c3_max
  R32 = c3_A2/c3_max
  R33 = c3_A3/c3_max
  R34 = c3_A4/c3_max
  R35 = c3_A5/c3_max

  #mencari nilai normalisasi R4
  R41 = c4_min/c1_A4
  R42 = c4_min/c2_A4
  R43 = c4_min/c3_A4
  R44 = c4_min/c4_A4
  R45 = c4_min/c5_A4

  #mencari nilai normalisasi R5
  R51 = c5_min/c1_A5
  R52 = c5_min/c2_A5
  R53 = c5_min/c3_A5
  R54 = c5_min/c4_A5
  R55 = c5_min/c5_A5

  #menampilkan hasil Normalisasi
  print('-'*20)
  df_R = pd.DataFrame({'R':['R1','R2','R3','R4','R5'],
                      '1':[R11,R12,R13,R14,R15],
                      '2':[R21,R22,R23,R24,R25],
                      '3':[R31,R32,R33,R34,R35],
                      '4':[R41,R42,R43,R44,R45],
                      '5':[R51,R52,R53,R54,R55]})
  df_R1 = df_R.set_index(df_R['R']).drop(columns=['R'])
  print('Hasil Normalisasi')
  print(df_R1)
  print('-'*20)

  #mencari nilai V
  v1 = w1*R11+w2*R12+w3*R13+w4*R14+w5*R15
  v2 = w1*R21+w2*R22+w3*R23+w4*R24+w5*R25
  v3 = w1*R31+w2*R32+w3*R33+w4*R34+w5*R35
  v4 = w1*R41+w2*R42+w3*R43+w4*R44+w5*R45
  v5 = w1*R51+w2*R52+w3*R53+w4*R54+w5*R55

  #menampilkan nilai V
  print('-'*20)
  print('Nilai V1:', v1)
  print('Nilai V2:', v2)
  print('Nilai V3:', v3)
  print('Nilai V4:', v4)
  print('Nilai V5:', v5)
  print('-'*20)

  #membuat plot
  import matplotlib.pyplot as plt

  data = {'V1':v1,'V2':v2,'V3':v3,'V4':v4,'V5':v5}
  courses = list(data.keys())
  values = list(data.values())
  
  fig = plt.figure(figsize = (10, 5))
  plt.bar(courses, values, color ='maroon',width = 0.4)
  plt.xlabel("V")
  plt.ylabel("Nilai V")
  plt.title("Barplot Calon Pegawai Terbaik dengan nilai V")
  plt.show()

  #membuat IF Else dan menentukan calon pegawai terbaik
  if (v1>v2) and (v1>v3) and (v1>v4) and (v1>v5):
    print('Nilai Terbesar adalah :',v1)
    print('Dan')
    print('Calon Pegawai terbaik adalah Calon Pegawai Pertama')
  
  elif (v2>v1) and (v2>v3) and (v2>v4) and (v2>v5):
    print('Nilai Terbesar adalah :',v2)
    print('Dan')
    print('Calon Pegawai terbaik adalah Calon Pegawai Kedua')

  elif (v3>v1) and (v3>v2) and (v3>v4) and (v3>v5):
    print('Nilai Terbesar adalah :',v3)
    print('Dan')
    print('Calon Pegawai terbaik adalah Calon Pegawai Ketiga')

  elif (v4>v1) and (v4>v2) and (v4>v3) and (v4>v5):
    print('Nilai Terbesar adalah :',v4)
    print('Dan')
    print('Calon Pegawai terbaik adalah Calon Pegawai Keempat')

  elif (v5>v1) and (v5>v2) and (v5>v3) and (v5>v4):
    print('Nilai Terbesar adalah :',v5)
    print('Dan')
    print('Calon Pegawai terbaik adalah Calon Pegawai Kelima')
```
