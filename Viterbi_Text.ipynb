#**CASE FOLDING**

```python
#membuat fungsi case folding
import re, string, unicodedata 
def Case_Folding(text):
    # Hapus non-ascii
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    # Menghapus Tanda Baca
    text = re.sub(r'[^\w]|_',' ', text)
    
    # Menghapus Angka
    text = re.sub("\S*\d\S*", "", text).strip()
    text = re.sub(r"\b\d+\b", " ", text)
    
    # Menghapus white space
    text = re.sub('[\s]+', ' ', text)
    
    return text
```

```python
# get the tags counts to a variable for easy access to total tag count for a tag
tag_cnt = Counter(tags)
tag_cnt 
```

```python
word_given_tags = pd.Series(FreqDist(train_tokens)).reset_index()\
                                .rename(columns={'level_0':'vocabulary','level_1':'tag',0:'count'})\
                                .reset_index(drop=True).sort_values(by='count',ascending=False)
word_given_tags
```

```python
# method to compute word given tag: count
def word_given_tag(word, tag):
    word_given_tag_count = word_given_tags.loc[(word_given_tags['vocabulary'] == word) & (word_given_tags['tag'] == tag),'count']    
    count = word_given_tag_count.to_list()
    return word_given_tag_count.to_list()[0] if len(count) != 0 else 0
```

```python
# compute occurance of a tag1 followed by tag2: tag2(t2) given tag1 (t1), i.e. Transition Probability
def t2_given_t1(t2, t1, train_bag = train_tokens):
    
    # get tag count for t1
    count_t1 = tag_cnt[t1]
    count_t2_t1 = 0
    
    # iterate through all the tags
    for index in range(len(tags)-1):
        
        # check if t1 is followed by t2 in the on each iterated tag index
        if tags[index]==t1 and tags[index+1] == t2:
            count_t2_t1 += 1
            
    return (count_t2_t1, count_t1) 
```

##**MATRIKS EMISI**

```python
# convetr the above dataframe to pivot table with index as vocaublary and colunms as tag highlighting the probability as values
emission_probability = pd.pivot_table(word_given_tags,columns='tag',index='vocabulary',values='count',fill_value=0.0)
emission_probability = emission_probability.apply(lambda x: x/x.sum(),axis=0)
emission_probability
```

##**MATRIKS TRANSISI**

```python
# creating t x t transition matrix of tags
# each column is t2, each row is t1
# thus M(i, j) represents P(tj given ti)

# initialize txt matrix with zeroes
tags_matrix = np.zeros((len(tag_set), len(tag_set)), dtype='float32')

# get the transition probability of each tag followin another tag
for i, t1 in enumerate(list(tag_set)):
    for j, t2 in enumerate(list(tag_set)): 
        tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/tag_cnt[t2]
```

```python
transition_probability = pd.DataFrame(tags_matrix, columns = list(tag_set), index=list(tag_set))
transition_probability
```

##**MEMBANGUN VITERBI**

```python
# created a vairbale to view all the unknown words for prediction
unknow_words = set()

# Viterbi
def Viterbi(words, train_bag = train_tokens):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        for tag in T:
            
            # for the first tage the tag t2 will always be 'fw'
            if key == 0:
                transition_p = transition_probability.loc['fw', tag]
            else:
                transition_p = transition_probability.loc[state[-1], tag]
                
            # compute emission probabilities
            if word in vocabulary:
                emission_p = emission_probability.loc[word,tag]
            else:
                unknow_words.add(word)
                # emission probability will be 0 for unknown
                emission_p = 0 
            # state probability of Viterbi is emiision probability * transition probability
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
        pmax = max(p)
        # getting state for which probability is maximum
        state_max = T[p.index(pmax)]
        state.append(state_max)
    return list(zip(words, state))
```
